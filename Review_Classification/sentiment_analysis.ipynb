# 아이스크림

```python
DF = pd.read_csv('./data/아이스크림 연수후기.csv').dropna()
DF2 = DF.rename(columns = {'0': 'class', '1': 'title', '2': 'rated', '3': 'comments', '4': 'classified', '5': 'date'})
```

```python
DF2 = DF2.reset_index(drop=True)
```

```python
DF2['title_comments'] = DF2['title'] + ' ' +  DF2['comments']
```

```python
DF2.to_csv('icecream.csv', index = False)
```

```python
DF2
```

```python
DF3 = DF2[DF2['title_comments'].apply(lambda x: False if len(x) > 512 else True)]
```

```python
DF3 = DF3.reset_index(drop=True)
```

```python
DF3
```

```python
comment_list = DF3['title_comments']
comment_list
```

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="matthewburke/korean_sentiment")

emot_list = []

for i in range(0, len(comment_list)):
    preds = classifier(comment_list[i], return_all_scores=True)
    is_positive = preds[0][1]['score'] > 0.5
    emot_list.append(is_positive)
```

```python
emot_list
```

```python
DF4 = DF3.copy()

DF4['emotion_analysis'] = emot_list
```

```python
DF4.to_csv('icecream_all_emotion.csv', index=False)
```
